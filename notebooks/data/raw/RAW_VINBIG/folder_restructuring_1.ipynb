```python
from pathlib import Path
import shutil

# Combine all VINBIG image shards into one folder.
# Expected layout: RAW_VINBIG/train/images, ..., RAW_VINBIG/test/images

raw_nih_dir = Path.cwd()
if raw_nih_dir.name != "RAW_VINBIG":
    raw_nih_dir = Path("notebooks/data/raw/RAW_VINBIG")

destination_dir = raw_nih_dir / "images"
destination_dir.mkdir(exist_ok=True)

copied = 0
skipped = 0

folders_to_check = ["train", "test"]  # change these to the folders you want

for folder_name in folders_to_check:
    source_dir = raw_nih_dir / folder_name

    if not source_dir.exists():
        print(f"Missing folder: {source_dir}")
        continue

    for image_path in source_dir.iterdir():
        if not image_path.is_file():
            continue

        destination_path = destination_dir / image_path.name
        if destination_path.exists():
            skipped += 1
            continue

        shutil.copy2(image_path, destination_path)
        copied += 1

print(f"Copied {copied:,} images into {destination_dir}")
print(f"Skipped {skipped:,} existing images")
```
